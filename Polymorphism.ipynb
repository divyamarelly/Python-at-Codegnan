```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old and I am a {self.gender}"
        
class Student(Person):
    def __init__(self, stdid, name, gender, age, branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch
    def details(self):
        return f"My name is {self.name} and I belongs to {self.branch} Branch"
        
s = Student(101, "Divya", "Female", 20, "AI&DS")
print(s.name)
print(s.branch)
print(s.details())
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name}, I am {self.age} years old and I am a {self.gender}"
        
class Student(Person):
    def __init__(self, stdid, name, gender, age, branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch
    def details(self):
        print(super().details())
        return f"My name is {self.name} and I belongs to {self.branch} Branch"
        
s = Student(101, "Divya", "Female", 20, "AI&DS")
print(s.details())
```

```python
# Method Overriding 
# Even if you define different number of parameters in methods, always child method overrides parent class methods
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self, location):
        return f"My name is {self.name} I am staying in {location}"
        
class Student(Person):
    def __init__(self, stdid, name, gender, age, branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch
    def details(self):
        return f"My name is {self.name} and I belongs to {self.branch} Branch"
        
s = Student(101, "Divya", "Female", 20, "AI&DS")
print(s.details())
print(s.details("India"))
```

```python
# Method Overloading
# Directly Python doesn't support method overloading but, we can achieve same functionality through the default arguments or Arbitary aruguments
class Math:
    def add(self,a,b,c=0,d=0):
        return a + b + c + d
m=Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```

```python
# Arithmetic Operators using magic methods
class Point:
    def __init__(self,x):
        self.x=x
    def __str__(self): # String magic method
        return f"This is Point class and Point value is {self.x}"
    def __add__(self, other):
        return self.x + other.x
    def __sub__(self, other):
        return self.x - other.x
    def __mul__(self, other):
        return self.x * other.x
    def __floordiv__(self, other):
        return self.x // other.x
    def __mod__(self, other):
        return self.x % other.x
p1 = Point(5)
p2 = Point(3)
print(p1)
print(p2)
p3 = p1 + p2
p4 = p1 - p2
p5 = p1 * p2
p6 = p1 // p2
p7 = p1 % p2
print(p3)
print(p4)
print(p5)
print(p6)
print(p7)
```

```python
# Relational Operaotrs using magic methods
class Point:
    def __init__(self, x):
        self.x = x
    def __str__(self):
        return f"Point value is {self.x}"
    # Relational Operators
    def __lt__(self, other):      
        return self.x < other.x
    def __gt__(self, other):     
        return self.x > other.x
    def __le__(self, other):      
        return self.x <= other.x
    def __ge__(self, other):      
        return self.x >= other.x
    def __eq__(self, other):     
        return self.x == other.x
    def __ne__(self, other):     
        return self.x != other.x

p1 = Point(5)
p2 = Point(3)
print(p1)
print(p2)
print(p1 < p2)
print(p1 > p2)
print(p1 <= p2)
print(p1 >= p2)
print(p1 == p2)
print(p1 != p2)
```
