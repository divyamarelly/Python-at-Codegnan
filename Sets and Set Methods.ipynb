### Set Characteristics

```python
# Set is unordered collection
# Set returns unique elements
s = {1,2,3,2,3,'hi','a',2.02,(10,20,30)}
print(s)
# print(s[0]) # set doesn't support indexing
```

```python
# list can't be used as set elements
s = {1,2,[3,2,3],'hi','a',2.02,(10,20,30)}
print(s)
```

### Set Methods

```python
# Element adding method
# add(ele): It adds the element to set if element not in set, otherwise nothing
s = {1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("Code")
print(s)
```

```python
# Element removing methods
# remove(item): It removes the item if it exists in set otherise 'KeyError'
# discard(item): It removes the item if it exists in set otherwise nothing
s={1,2,3,4,5}
s.remove(5)
print(s)
# s.remove(100) # KeyError
s.discard(3)
print(s)
s.discard(100)
print(s)
```

### Set Operations

```python
# Sets Operations
# union(set): It returns the all unique elements from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
union=s1.union(s2,s3)
print(union)
res= s1 | s2 | s3
print(res)
```

```python
# Sets Operations
# intersection(set): It returns the all common elements from all sets
s1={1,3,5,7}
s2={1,4,5,6,7}
s3={1,2,7,8,9}
intersection=s1.intersection(s2,s3)
print(intersection)
res= s1 & s2 & s3
print(res)
print(s1,s2,s3)
```

```python
# Sets Operations
# intersection_update(set)
s2={1,4,5,6,7}
s3={1,2,7,8,9}
s1.intersection_update(s2,s3)
print(s1,s2,s3)
```

```python
# Sets Operations
# difference(set): It returns a set of elements those are not in another set
s1={1,3,5,7,10}
s2={4,5,6,7}
s3={1,2,7,8,9}
difference=s1.difference(s2,s3)
print(difference)
res= s1 - s2 - s3
print(res)
print((s1 | s2) - s3)
```

```python
# Sets Operations
# symmetric_difference(set): It returns the non common elements between all sets
s1={1,3,5,7,10}
s2={4,5,6,7}
s3={1,2,7,8,9}
symmetric_difference=s1.symmetric_difference(s2)
print(symmetric_difference)
res= s1 ^ s2 ^ s3
print(res)
```

```python
# issubset(set)
# issuperset(set)
# isdisjoint(set)
s1={1,3,5,7,10}
s2={5,7}
s3={4,6}
print(s2.issubset(s1))
print(s1.issuperset(s3))
print(s3.isdisjoint(s1))
```
