```python
# Validate Aadhaar Number
import re
try:
    aadharno = input("Enter Aadhaar Number: ")
    res = re.fullmatch(r"^[2-9][0-9]{3}\s[0-9]{4}\s[0-9]{4}$", aadharno)
    print(res)
    if res:
        print(f"{aadharno} is valid")
    else:
        print(f"{aadharno} is invalid")
except:
    print("There is no matched pattern")
```

```python
# validate email id
import re
try:
    email = input()
    res = re.fullmatch(r"^[a-z0-9][a-z0-9.]+\@[a-z0-9]+\.[a-z]{2,3}$",email)
    print(res)
    if res:
        print(f"{email} is valid")
    else:
        print(f"{email} is invalid")
except:
    print("There is no matched pattern")
```

```python
# validate Date
import re
try:
    date = input()
    res = re.fullmatch(r"^(0?[1-9]|[12][0-9]|3[01])/(0?[1-9]|1[0-2])/(\d{4})",date)
    # print(res)
    if res:
        print(f"{date} is valid")
    else:
        print(f"{date} is invalid")
except:
    print("There is no matched pattern")
```
