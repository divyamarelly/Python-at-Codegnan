```python
import math
print(math.ceil(4.25)) # It returns the Upperbound integer value
print(math.floor(4.25)) # It returns the Lowerbound integer value
print(math.factorial(5)) # It returns the Factorial of a number
print(math.gcd(10, 6, 4)) # It returns the Greatest Common Divisor
print(math.lcm(10, 6, 4)) # It returns the Least Common Multiple
print(math.pow(10, 3)) # Pow(a, b) : It returns the Float tyoe of a raise to b
print(math.sqrt(9)) # It returns the square root value of a number
print(math.log10(2)) # To Find the Log10 values
```

```python
import random
print(random.random()) # This function generates the random number between 0 to 1
print(random.randint(1,6)) # This function generates the random number between a range 
print(random.randint(1000, 9999)) # 4 digit OTP generation
print(random.choice(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'])) # This function returns the random value from sequence
print(random.choices(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'], k=13))
arr = ['A',1,2,3,4,5,6,7,8,9,'J','Q','K']
random.shuffle(arr)
print(arr)
```

```python
# Who will reach the Top(100) in less steps
import random
rolls = count = 0
print("Welcome to the Who will Reach Top in Less Steps Game")
while True:
    input("To Throw a dice, please click enter button")
    face_val = random.randint(1,6)
    if count + face_val <= 100:
        count += face_val
    rolls += 1
    if count == 100:
        print(f"Your Dice value: {face_val}")
        print(f"You reached the Top for {rolls} chances")
        break
    print(f"Your Dice value: {face_val}")
    print(f"You are at {count} Position in {rolls} chances")
    print("*" * 50)
```
