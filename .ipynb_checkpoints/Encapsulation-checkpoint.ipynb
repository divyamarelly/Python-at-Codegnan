#### Encapsulation: 
    - Encapsulation means, wraping the methods and attributes in single unit that one is class. The main aim of Encapsulation is to restrict the direct access of attributes and methods.
    - This type of restriction can implement by using `Access Modifiers`
    - Types of Access Modifiers:
        1. Public
        2. Protected
        3. Private
    - In Python we don't have any keywords to represent it as public, protected and private, but by using underscore(_) we can define.
    1. Public: The Public Members(i.e attributes + methods) doesn't starts with under score. This public members can access any where in the program.
    2. Protected: The Protected Members(i.e attributes + methods) starts with single under score(_). This members can access in inside the class and in subclasses also.
    3. Private: The Private Members(i.e attributes + methods) starts with double under score(__). This members can access only inside the class.
    - Python doesn't support high security like Java. Here we can access the protected members and private menbers in outside the class by using the following ways:
    - To access the protected members: obj._ProtectedMemberName
    - To access the private members by using name mangling: obj._classname__PrivateMemberName
#### The Advantage of Encapsulation is:
    - It can provide the security for data and functionalities

```python
class Bank:
    def __init__(self, name, account, password, balance):
        self.name = name 
        self.account = account
        self.password = password
        self._balance = balance
    def __total_users(self): # getter method
        return "Total users : 101"
    def get_balance(self): # getter method
        return f"Current Balance is:{self._balance}"
    def deposit(self, deposit_amount): # setter method
        self._balance += deposit_amount
        return f"{deposit_amount} Deposited Successfully"
        
u1 = Bank("Divya",1001,"Divya@21",5000)
print(u1.name)
print(u1.get_balance())
print(u1.deposit(5000))
print(u1.get_balance())
print(u1._balance) # Protected members access in outside the class
print(u1._Bank__total_users()) # Name Mangling: syntax: obj._classname__attribute
```
