### Tuple Characteristics

```python
# characteristics
# Ordered Collection
t=(1,2,3,[10,30,40],23)
print(t)
```

```python
# immutable
t=(1,2,3,[10,30,40],23)
t[1]=100
print(t)
```

```python
# immutable
# After operation it doesn't change the length of object and doesn't change remaining elements position 
t=(1,2,3,[10,30,40],23)
t[3].append(100)
print(t)
```

```python
# Empty tuple representation
# Single element representation
t1=()
t2=tuple()
print(type(t1),type(t2))
```

```python
# Single element representation
t1=(10)
t2=(10,)
l1=[10]
print(type(t1),type(t2),type(l1))
```

```python
# Tuple Operations
# Tuple Concatenation
# Tuple Repetition
t1=(1,2,3)
t2=(3,4,5,6)
print(t1+t2)
print(t1*2)
print((0,)*10)
```

```python
# index and count()
# index(item): If item is present in list it returns the first occurence Index otherwise `ValueError` 
# count(item): It return the count of Total occurences of item 
t=(10, 20, 30, 20, 40, 50)
print(t.index(20))
print(t.count(20))
```

```python
# Built-in Functions- len(), min(), max(), sum()
t=(1, 2, 3, 3, 4, 5, 6)
print(len(t), min(t), max(t), sum(t))
print("Average:",sum(t)/len(t))
```
