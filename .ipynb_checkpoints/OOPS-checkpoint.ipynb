```python
# define Person class
class Student:
    school = "XYZ school"   # class attributes
    def __init__(self, name, age): #initialisation constructor(magic method)
    ## any methods starts and ends with dunderscore those methods are called as magic methods
        self.name = name  #Instance Attributes
        self.age = age     #Instance Attributes
        # print("This is Student class __init__method")
    def details(self):
        return f"My name is {self.name} and I am a {self.age} years old"

s1 = Student("Divya",20)
s2 = Student("Vamshi",22)
print(s1.name)
print(s2.age)
print(s1.details())
print(s2.details())
print(s1)
print(s1.school)
print(s2.school)
```
