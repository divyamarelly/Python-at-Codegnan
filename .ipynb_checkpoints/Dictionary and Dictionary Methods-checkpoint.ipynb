### Dictionary Characteristics

```python
# Dictionary ordered collection
details={"name" : "Divya", "course" : "Python", "batch" : 57, "age" : 20}
print(details)
```

```python
# Dictionary values access by using keys
# updating dictionary values
# adding items to dictionary
details={"name" : "Divya", "course" : "Python", "batch" : 57, "age" : 20}
print(details['course'])
# print(details['location']) # KeyError: 'location
# updating dictionry values
details['age']=21
print(details['age'])
#adding new items
details['location']="Hyderabad"
print(details)
```

```python
# adding age
details={"name" : "Divya", "course" : "Python", "batch" : 57, "age" : 20}
details['age']=details['age']+1
print(details)
```

```python
# dictionary keys are unique
details={"age" : 20, "name" : "Divya", "course" : "Python", "batch" : 57, "age" : 21, "location" : "Hyderabad"}
print(details)
```

```python
# list can't be used as dictionary keys
# dictionary keys are unique
details={"age" : 20, "name" : "Divya", "course" : "Python", [1,2,3] : 57, "age" : 21, "location" : "Hyderabad"}
print(details)
```

```python
details={"age" : 20, "name" : "Divya", "course" : "Python", "batch" : 57, "age" : 21, "location" : "Hyderabad", "marks" : [75,85,95]}
print(details)
```

### Dictionary Methods
- get(key, default_value:Optinal): It returns the value if key is exists in dictionary, otheriwse it returns default value.

```python
# dictionary keys are unique
details={"age" : 20, "name" : "Divya", "course" : "Python", "batch" : 57, "age" : 21, "location" : "Hyderabad"}
print(details["name"])
# print(details["marks"]) # This one raises KeyError
print(details.get("name"))
print(details.get("marks")) # Tget method by default it returns None if key not exists in dictionary
print(details.get("marks", "Marks not available")) # it returns default value 
```

- `update(new_dictionary)` : This method updates the values if new_dictionary keys already exists in Dictionary, otherwise it adds new items to the dictionary.

```python
details={"age" : 20, "name" : "Divya", "course" : "Python", "batch" : 57}
details.update({"batch" : 58, "course" : "Java", "location" : "hyderabad"})
print(details)
```

- `keys()` : This methods returns the list of dictionary keys
- `values()` : This methods returns the list of dictionary values
- `items()` : Method return the List of Tuple of key-value pairs

```python
details= {'age': 20, 'name': 'Divya', 'course': 'Python', 'batch': 57, 'location': 'hyderabad'}
keys=details.keys()
print(type(keys))
print(keys)
print(details.values())
print(details.items())
```

- popitem() : It return and removes the last item as tuple
- pop(key) : This method removes the specified key items and return the specified key's value, otherwise `KeyError`

```python
details= {'age': 20, 'name': 'Divya', 'course': 'Python', 'batch': 57, 'location': 'hyderabad'}
item=details.popitem()
print(item)
print(details)
val=details.pop('age')
print(val)
print(details)
```

```python
# Membership 
# It checks with dictionary keys
details= {'age': 20, 'name': 'Divya', 'course': 'Python', 'batch': 57, 'location': 'hyderabad'}
print('Divya' in details)
print('name' in details)
```
