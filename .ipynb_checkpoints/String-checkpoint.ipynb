```python
#indexing
s="Codegnan"
print(s[2])
print(s[-1])
```

```python
#slicing
s="Codegnan"
print(s[2:7])
print(s[-4:])
print(s[5:2])
print(s[3:100])
print(s[::-1])
print(s[-3::-1])
print(s[-3::1])
```

```python
#Empty String Representation
#type() : This function returns the objects datatype
s1=""
s2=str()
print(type(s1))
print(type(s2))
```

```python
#String reading from user
#input() : This function is used to read dynamically from user and this function by default read it as string datatype
s=input()
print(type(s))
print(s)
```

```python
#String Operations
#1. String Concatenation (using + operator)
#2. Stirng Repetation (using * operator)
s1="Code "
s2="gnan"
s3=s1+s2
print(s3)
print(s1*3)
```

```python
#immutable
s="Codegnan"
res=s[:1]+"z"+s[2:]
print(res)
```
