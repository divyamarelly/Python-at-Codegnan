```python
# return 1 to n numbers using functions
import sys
def fun(n):
    res = []
    for i in range(1, n+1):
        res.append(i)
    return res
r = fun(10)
print(sys.getsizeof(r))
print(r)
```

```python
# return 1 to n numbers using generator function
import sys
def gen(n):
    for i in range(1, n+1):
        yield i
r = gen(3)
print(r)
# n = next(r)
# print(sys.getsizeof(n))
print(next(r))
print(next(r))
print(next(r))
print(next(r)) # StopIteration, Because there is no element left to get value in generator object
```

```python
# return 1 to n numbers using generator function
import sys
def gen(n):
    res = []
    for i in range(1, n+1):
        yield i
r = gen(3)
print(r)
# to prevent the StopIteration error by using for loop
for val in r:
    print(val)
```

```python
# return 1 to 100 cubic values using functions
import sys
import time
s = time.time()
def fun(n):
    res = []
    for i in range(1, n+1):
        res.append(i**3)
    return res
r = fun(100)
print(sys.getsizeof(r))
print(r)
e = time.time()
print(e-s)
```

```python
# return 1 to 100 cubic values using generator function
import time
import sys
s = time.time()
def gen(n):
    for i in range(1, n+1):
        yield i ** 3
r = gen(100)
print(r)
# to prevent the StopIteration error by using for loop
for val in r:
    print(val, end=" ")
print()
e = time.time()
print(e-s)
```
