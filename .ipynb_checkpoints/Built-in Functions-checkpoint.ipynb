```python
# chr() : This function return the character based on ASCII value
# ord() : It returns the ASCII value based on character
print(chr(65))
print(ord('A'))
print(chr(35))
print(ord('1'))
```

```python
# abs() : It returns the Absolute value
print(abs(10-20))
```

```python
# find the difference between two strings
s1 = "CODEA"
s2 = "GNANA"
n = len(s1)
s = ""
for i in range(n):
    diff = abs(ord(s1[i]) - ord(s2[i]))
    if diff == 0:
        s += s1[i]
    else:
        s += chr(64+diff)
print(s)
```

```python
# find bigger and smaller characters from s1 and s2
# compare each character in both strings and if s1[i] > s2[i] add s1[i] to bigger and s2[i] to smaller
# if both same leave it
s1 ,s2 = map(str, input().upper().split())
n = len(s1)
big = small = ""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i] > s2[i]:
            big += s1[i]
            small += s2[i]
        else:
            big += s2[i]
            small += s1[i]
print(big)
print(small)
```

```python
# map(fun, iterable): It applies the functionality on each element in iterable
# filter(fun, iterable): It filters the iterables, it returns the values which is not None
# find the square of list of integers
def square(x):
    return x**2
lst = [1,2,3,4,5]
print(list(map(square, lst)))
```

```python
# filter the even numbers from list of integers
def even(x):
    if x % 2 == 0:
        return x
arr = [1,2,3,4,5,6,7]
print(list(filter(even, arr)))
print(list(map(even, arr)))
```

```python
def sum(t):
    x, y = t
    return x + y
lst = [(1,2),(3,4)]
print(list(map(sum, lst)))
```

```python
# zip(iterable1, iterable2, iterable3, ....., iterableN):
# It returns list of tuples, the tuple is generated as each position elements from all tuples
name = ['Divya', 'Pranathi', 'Susmitha', 'Priya', 'Rishitha', 'Sravanthi']
age = [20,21,21,21,21,21]
print(list(zip(name, age)))
```

```python
# enumerate(iterable): it returns the list of tuple of index and value
arr = [10,20,30]
print(list(enumerate(arr)))
```

```python
# find the target element in list
arr = [10, 20, 30, 40, 50]
target = 40
for ind, val in enumerate(arr):
    if target == val:
        print(ind)
        break
else:
    print(-1)
```

```python
# any(iterable): It returns True when any one iterable element's boolean value is True
arr = [1,2,3,4,0,0,0]
print(any(arr))
print(any([0,0,0,0]))
# all(iterable): It returns True when all iterable element's boolean value is True
print(all([1,2,3,4]))
print(all([True, False, True]))
print(all([False, False, False]))
```

```python
# isinstance(obj, class_name): It checks whether the object is belongs to particular class or not
print(isinstance(10, int))
print(isinstance(12.25, int))
```

```python
# sorted(): it returns the new sorted object, by default it returns the descending order
# reversed(): it returns the reverse of an object value
arr = [10,2,34,35,3,5,6]
res = sorted(arr, reverse = True)
print(res)
print(arr)
print(list(reversed(arr)))
print("".join(list(reversed("Hello"))))
```

```python
arr = ["Hello", "Hi", "apple", "banana"]
sorted(arr, key = len)
```

```python
# count the fequency of characters in string
string = "Hello"
frequency = {}
for char in string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print(frequency)
```
