```python
# scopes of variable
# 1. Local scope
# 2. Enclosing scope
# 3. Global scope
# 4. Builtin scope
x = 10   # global scope variable
def outer():
    y = 20   # Enclosing scope variable
    def inner():
        z = 30   # local scope variable
        print(z, y, x)   # print() - Builtin scope function
    inner()
outer()
```

```python
# using global scope variable in local scope
x = 100
def outer():
    y = x + 1
    print("Local Scope Variable Value:", y)
outer()
print("Global Scope Varibale Value:", x)
```

```python
# Updating the Global Scope Variable Value in Local Scope
# can't update global scope variable in local scope directly 
x = 100
def outer():
    x = x + 1
    print("Local Scope Variable Value:", x)
outer()
print("Global Scope Varibale Value:", x)
```

```python
# Updating the Global Scope Variable Value in Local Scope by using 'global' Keyword
x = 100
def outer():
    global x
    x = x + 1
    print("Local Scope Variable Value:", x)
outer()
print("Global Scope Varibale Value:", x)
```

```python
# Updating the Enclosing scope variable value in local scope
x = 10   # global scope variable
def outer():
    y = 20   # Enclosing scope variable
    def inner():
        z = 30 
        y = y + z   # local scope variable
        print(z, y, x)   # print() - Builtin scope function
    inner()
outer()
```

```python
# Updating the Enclosing scope variable value in local scope
x = 10   # global scope variable
def outer():
    y = 20   # Enclosing scope variable
    def inner():
        nonlocal y
        z = 30 
        y = y + z   # local scope variable
        print(z, y, x)   # print() - Builtin scope function
    inner()
    print("Enclosing scope variable value:", y) 
outer()
```

```python
x = 100
def outer():
    x = 200
    x = x + 2
    print("Local scope variable x value:", x)
outer()
print("Global scope variable x value:", x)
```
