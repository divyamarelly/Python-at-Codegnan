- ### Exception Handling

- `EXCEPTION`: It is nothing but it is an error.
              - Whenever an execution is occurs it stops the program execution and it returns type of exception and followed by execution messsage. 

-`Types Of Error:` <br>
    1. `Syntax Error (or) Compiletime Error` : The syntax error raises during the compilation time of the program, those errors are called as `Syntax Error`. <br>
    `EX` : Intendationerror, Syntaxerror <br>
    2. `Runtime Error` : The runtime error raises during Runtime of a program, those errors are called as `Runtime Error`  <br>
                        - We cannot handle syntax error but we can handle the Runtime Errors by using Exception handling blocks(ie, try-except) <br>
  `EXCEPTION HANDLING BLOCKS`: <br>
                    `Syntax`: <br>
                            try: <br>
                                # block code that might contains exception  <br>
                            except: <br>
                                # exception handling block <br>
                            else: [optional]<br> 
                                # This block executes there is no exception raised in try block <br>
                            finally: [optional] <br>
                                # This block always executes every exception raised or not. <br>
                               
                        
`EX` : KeyError, ValueError, Zero Division Error.  <br>
    3. `Logical Error` :  <br>

```python
try:
    list = list[1,2,3]
    print(list[10])
    print("Hi")
except:
    print("Something wrong in try block")
else: 
    print("No exception raised in try block")
finally:
    print("Program done")
```

```python
try:
    list = [1,2,3]
    print(list[2])
    print("Hi")
except:
    print("Something wrong in try block")
else: 
    print("No exception raised in try block")
finally:
    print("Program done")
```

```python
# we cannot handle the syntax error
try:
    print("hello')
except:
    print("something wrong in try block")
else: 
    print("no exception raised in try block")
finally:
    print("program done")
```

```python
try:
    a = 10
    b = 0
    c = a/b
    print(c)
except Exception as E:
    print(f"something wrong in try block: {E}")
```

```python
try:
    a = 10
    b = a.append(0)
    c = a/b
    print(b)
except Exception as E:
    print(f"something wrong in try block: {E}")
```

```python
try:
    a = 10
    b = a.append(0)
    print(b)
except ZeroDivisionError as e:
    print(f"Zero division not possible: {e}")
except ValueError as e:
    print(f"Value Error raised in Try block: {e}")
          
#except Exception as e:
 #   print(f"something wrong in try block: {e}")
```

```python
try:
    a = 10
    b = 0
    c = a/b
    print(c)
except ZeroDivisionError as e:
    print(f"Zero division not possible: {e}")
except ValueError as e:
    print(f"Value Error raised in Try block: {e}")
          
except Exception as e:
    print(f"something wrong in try block: {e}")
```

```python
# It executes the specific exception block when specified exception raised otherwise, it executes the Common except block
try:
    a = 10
    b = a.append(1)
    c = a/b
    print(c)
except ZeroDivisionError as e:
    print(f"Zero division not possible: {e}")
except ValueError as e:
    print(f"Value Error raised in Try block: {e}")
except Exception as e: # Common except block with Exception class
    print(f"something wrong in addition.add: {e}")
```
