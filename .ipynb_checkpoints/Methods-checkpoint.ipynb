- Methods: The Function definition is implemented inside the class those functions
called as Methods.
    - Types of Methods:
      1. Instance Method
      2. Class Method
      3. Static Method

```python
# class attribute - school
# instance attribute - id, name, class
class Student:
    school = "Codegnan School" 
    def __init__(self, stdid, name, standard):
        self.stdid = stdid
        self.name = name
        self.standard = standard
    def student_details(self):
        return f"My self {self.name} and I am from {self.standard} class"
s = Student(340, "Divya", 9)
print(s.name)
print(s.school)
print(s.student_details())
```

### Methods
- Instance Method: The method definition takes first parameter as 'self' thoes methods
called as `Instance Method`. This methods works on insatnce attributes. <br>
 `syntax:` <br>
    `            class class_name:`   <br>
    `               def method_name(self):`  <br>
    `                   #block code`  <br>
- Class Method: The method definition starts with `classmethod` decorator and it takes first parameter s `cls` those methods called as `Class Methods`. This methods works on class Attributes.
  `syntax:` <br>
    `            class class_name:`  <br>
    `                @classmethod`   <br>
    `               def method_name(self):`  <br>
    `                   #block code`  <br>
- Static Method: The method definition starts with `staticmethod` decorator and it doesn't take first parameter as `self or cls` those methods called as `Static methods`. This methods works on except instance attributes and class attributes.
  `syntax:` <br>
    `            class class_name:`  <br>
    `                @staticmethod`   <br>
    `               def method_name(parameters):`  <br>
    `                   #block code`  <br>
- Initialization Constructor (__init__): The initialization constructor is used to initialize the instance attributes data and this constructor takes first parameter as self and this constructor executes whenever an object is created for that class.

```python
# Instance Method
# class attribute - school
# instance attribute - id, name, class
class Student:
    school = "Codegnan School" # class attributes
    def __init__(self, stdid, name, standard):
        self.stdid = stdid # instance attributes
        self.name = name # instance attributes
        self.standard = standard # instance attributes
    def student_details(self): # Instance Method
        return f"My self {self.name} and I am from {self.standard} class"
    # class Method
    @classmethod
    def update_school(cls, new_school_name):
        cls.school = new_school_name
        return "School name Updated"
    # static Method
    @staticmethod
    def percentage(marks:list):
        return (sum(marks)/300) * 100
        
s1 = Student(101, "Divya", 9)
s2 = Student(102, "Vamshi", 10)
print(s1.school, s2.school)
print(s.student_details())
print(s1.update_school("CODEEDGE school"))
print(s1.school, s2.school)
print(s2.percentage([75,80,85]))
```
