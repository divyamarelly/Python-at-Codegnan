```python
n = 5
for i in range(n):
    for j in range(n):
        if i == j or j == n-1:
            print("*", end = " ")
        else:
            print("*", end = " ")
    print()
```

```python
n = 5
k = 1
for i in range(n):
    for j in range(n):
        if j >= n - i - 1:
            print(k, end = "\t")
            k += 1
        else:
            print(" ", end = "\t")
    print()
```

```python
n = 7
for i in range(n):
    for j in range(n):
        if j == 0 or j == n - 1 or (i == j and i <= n//2) or (i <= n//2 and j == n - i - 1) :
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print()
```

```python
n = 5
for i in range(n):
    for j in range(n):
        if i % 2 == 0:
            print(i * n + j + 1, end = "\t")
        else:
            print((i + 1) * n - j, end = "\t")
    print()
```

```python
n = 5
k = n * (n + 1) // 2
for i in range(n):
    for j in range(n):
        if i >= j:
            print(k, end = "\t")
            k -= 1
        else:
            print(" ", end = "\t")
    print()
```

```python
n = 8
k = 0

for i in range(n):
    for j in range(n):
        if i >= j:
            print(chr(65 + (k % 26)), end=" ")
            k += 1
        else:
            print(" ", end=" ")
    print()
```
