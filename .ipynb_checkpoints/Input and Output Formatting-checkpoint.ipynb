### Input Formatting

```python
# reading integer from user
num = int(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# reading float value from user
val = float(input())
print(type(val))
print(val)
```

```python
# complex reading from user
val = complex(int(input()), int(input()))
print(type(val))
print(val)
```

```python
# boolean reading as integer
val = bool(int(input()))
print(type(val))
print(val)
```

```python
# reading string value from user
s = input()
print(type(s))
print(s)
```

```python
# reading list of integers from users
lst = input().split()
print(lst)
```

```python
# reading list of integers from users
lst = map(int, input().split())
print(lst)
```

```python
# reading list of integers from users
lst = list(map(int, input().split()))
print(lst)
```

```python
# reading list of string from user
s = list(map(str, input().split()))
print(s)
```

```python
# approach 2
s = input().split()
print(s)
```

```python
# reading tuple of integers from user
t = tuple(map(int, input().split()))
print(t)
```

```python
# reading set of integers from user
s = set(map(int, input().split()))
print(s)
```

```python
# eval(): This function evaluates the string type expression
val = eval(input()) # this evaluates all data types
print(type(val))
print(val)
```

```python
print("3 + 5")
print(eval("3 + 5"))
```

### Output Formatting

```python
# using string concatenation
name="Divya"
age=20
# My name is Divya and I am 20 years old
print("My name is "+name+" and I am "+str(age)+" years old")
```

```python
# 1. String formatting using comma(old style format)
name="Divya"
age=20
# My name is Divya and I am 20 years old
print("My name is",name,"and I am",age,"years old")
```

```python
# 2. Access Specifiers
# string - %s
# integer - %d
# float - %f
name = "Divya"
age = 20
# My name is Divya and I am 20 years old
print("My name is %s and I am %d years old"%(name,age))
```

```python
name = "Divya"
age = 20
# My name is Divya and I am 20 years old
print("My name is %s and I am %d years old"%(age,name))
```

```python
# 3. Dot Format
name = "Divya"
age = 20
# My name is Divya and I am 20 years old
print("My name is {} and I am {} years old".format(name,age))
print("My name is {} and I am {} years old".format(age,name))
```

```python
# 4. F-Format
name = "Divya"
age = 20
# My name is Divya and I am 20 years old
print(f"My name is {name} and I am {age} years old")
print(f"My name is {age} and I am {name} years old")
```
