```python
def func(x):
    if x == 0:
        return 0
    x = x - 1
    func(x)
    print(x)
func(3)
```

```python
# print 1 to n using recursion
def func(x):
    if x == 0:
        return 0
    func(x-1)
    print(x)
print(func(3))
```

```python
# print n to 1 using recursion
def func(x):
    if x == 0:
        return 0
    print(x)
    func(x-1)
func(5)
```

```python
# Factorial of a number
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
# sum of digits in a number
# n = 456 o/p: 15
def sumofDigits(n):
    if n == 0:
        return 0
    return n % 10 + sumofDigits(n//10)
print(sumofDigits(456))
print(sumofDigits(15346))
```

```python
# count digits in a number
# n = 456 o/p: 3
def sumofDigits(n):
    if n == 0:
        return 0
    return 1 + sumofDigits(n//10)
print(sumofDigits(456))
print(sumofDigits(15346))
```

```python
# find the square of a number using lambda
s = lambda x:x**2
print(s(10))
```

```python
# find the square of list of integers using map and lambda
arr = [1,2,3,4,5]
print(list(map(lambda x: x**2, arr)))
```

```python
arr = [1,2,3,4,5]
print(list(filter(lambda x: x%2==0, arr)))
```

```python
n = 10
res = "Even" if n % 2 == 0 else "Odd"
print(res)
```

```python
# check whether the number is even or odd using lambda
even_or_odd = lambda x:"Even" if x%2==0 else "Odd"
print(even_or_odd(10))
```

```python
arr = [1,2,3,4,5]
# 1 - odd
# 2 - even
arr = [1,2,3,4,5]
print(list(map(lambda x:f"{x} - even" if x%2==0 else f"{x} - odd",arr)))
```

```python
# find the frequency of each character
string = input()
count = {}
for char in string:
    if char in count:
        count[char] += 1
    else:
        count[char] = 1
print(count)
```

```python
# find the frequency of each character
string = input()
count = {}
for char in string:
    count[char] = count.get(char, 0) + 1
print(count)
```
