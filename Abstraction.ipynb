```python
from abc import ABC, abstractmethod
class Mobile(ABC):
    @abstractmethod
    def ram(self):
        pass
    @abstractmethod
    def rom(self):
        pass
    @abstractmethod
    def battery(self):
        pass
    # concreate method implementation
    def display(self, Hzs):
        return f"This mobile display quality is {Hzs} Hzs"
class Oppo(Mobile):
    def __init__(self, model, battery_size, display_quality):
        self.model = model
        self.battery_size = battery_size
        self.display_quality = display_quality
    def ram(self, ram_size):
        return f"This Oppo mobile {self.model} has {ram_size} GB RAM"
    def rom(self, rom_size):
        return f"This Oppo mobile {self.model} has {rom_size} GB ROM"
    def battery(self):
        return f"This Oppo mobile {self.model} has {self.battery_size} mah Battery Capacity"
    def camera(self, primary, secondary):
        return f"The Oppo {self.model} mobile have primary camera with {primary} MP and seconday camera with {secondary} MP"
# m = Mobile()  TypeError: Can't instantiate abstract class Mobile without an implementation for abstract methods 'battery', 'ram', 'rom'
reno15 = Oppo("reno15", 5000, 120)
print(reno15.ram(4))
print(reno15.rom(128))
print(reno15.battery())
print(reno15.camera(32, 52))
print(reno15.display(120))
```
