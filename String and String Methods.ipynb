## String Methods
- Case Conversion Methods
   - upper(): It converts entire string into Uppercase
   - lower(): It converts entire string into Lowercase
   - title(): It converts the first letter of every word into upper and remaining lower
   - swapcase(): It converts Uppercase letter into lowercase and lowercase to uppercase
   - capitalize(): It converts First character into Uppercase and remaining into lowercase

```python
string="The Sky is blue and Ocean is Blue"
print(string.upper())
print(string)
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- Remove white space or pattern at ends of String
   - strip(): By default it removes the white space at both ends of string
   - lstrip(): It removes the white space at left side of string
   - rstrip(): It removes the white space at right side of string

```python
s="     Codegnan      "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@@@Codegnan@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
#split(): It splits the entire string into list of substring based on specified seperator
#by default it splits at space
string="The Sky is blue and Ocean is Blue"
print(string.split())
print(string.split("is"))
print(string.split('e'))
```

```python
# 'sep'.join(iterable): It joins the all iterable elements with seperator
words=['The', 'Sky', 'is', 'blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
```

### Substring Index finding
- index(substring): If substrings is exists in string, it returns the first occurence of the substring position, otherwise `ValueError` 
- find(substring): If substrings is exists in string, it returns the first occurence of the substring position, otherwise `-1`

```python
string="The Sky is blue and Ocean is Blue"
print(string.index('is'))
print(string.index('z'))
```

```python
string="The Sky is blue and Ocean is Blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is',10)) #it start checking from 10th position
print(string.find('e',10))
print(string.find('e',-5))
```

```python
# count(substring): It returns the total non-overlapping occurences of substring
s="ABCBCBCbbaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('Z'))
```

```python
#replace(old_substring, new_substring): It replace the old sub-string with new sub-string in string
string="The Sky is blue and Ocean is Blue"
print(string.replace("blue","Red"))
print(string.replace('e','E'))
print(string.replace('e','E',2)) #it replace the first 2 occurences
```

### String Checking Methods
- *String checking methods returns the Boolean values as result*
- isupper(): It returns True, If entire string is in upper case
- islower(): It returns True, If entire string is in lower case
- isalpha(): It returns True, If entire string contains alphabets
- isdigit(): It returns True, If entire string contains digits
- isalnum(): It returns True, If entire string contains either alphabets or digits
- startswith(substring): It returns True, If string starts with substring
- endswith(substring): It returns True, If string ends with substring

```python
print("CODE".isupper())
print("CODE".islower())
print("code".islower())
print('1234'.isdigit())
print('123abc'.isdigit())
print("Code".isalpha())
print('Code123'.isalnum())
print('1234'.isalnum())
print("code".startswith('c'))
```
