```python
# generate total bill if total bill more than 500 then give 10% discount otherwise 0% discount
total_bill = int(input())
if total_bill > 500:
    discount = total_bill * 0.1
    print(f"Total bill after 10% discount: {total_bill - discount}")
else:
    print(f"Total bill after 0% discount: {total_bill}")
```

```python
# check whether the product stock is available or not (read product name and stock)
product = input()
stock = int(input())
if stock > 0:
    print(f"Currently {product} : {stock} items are available:")
else:
    print(f"Currently {product} out of stock")
```

```python
# A shop contains some items along with prices like {'pen':5, 'book':20, 'pencil':3}
# A customer want some items from shop, if item is available in shop, then find the total amount, otherwise Item not available
products = {'pen':5, 'book':20, 'pencil':3}
item = input().lower()
quantity = int(input())
if item in products:
    total_amount = products.get(item) * quantity
    print(f"Total amount of {quantity} {item} is: {total_amount}")
else:
    print(f"{item} not available")
```
