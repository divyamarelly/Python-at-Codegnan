### List Characteristics

```python
# Define List
# List is ordered Collection
lst=[1,2,3,4.5,"Hi","Hello",1,3,4]
print(lst)
```

```python
# List alllows indexing and Slicing
# Accessing List items
lst=[1,2,3,4.5,"Hi","Hello",1,3,4]
print(lst)
print(lst[5])
print(lst[-1])
print(lst[::2])
print(lst[-1:-5:-1])
print(lst[-4:10:1])
```

```python
lst=[10,20,30,40,50,60,70]
print(lst[:5:-1])
print(lst[:500])
print(lst[::-1]) # Reverse lsit
```

```python
# Empty list creation
l1=[]
l2=list()
print(type(l1),type(l2))
```

```python
# List Operations
# 1. List Concatenation (using + operator)
# 2. List Repetition (using * operator)
l1=[1,2,3]
l2=[3,4,5,6]
print(l1+l2)
print(l1*2)
print([0]*10)
```

### List Methods
- Item adding methods
   - append(item) : It adds item at end of list
   - extend(iterable) : It adds all iterator elements at end of list
   - insert(pos, item) : It adds item at specific position

```python
lst=[10,20,30,40]
lst.append(100)
print(lst)
lst.extend([1,2,3])
print(lst)
lst.append([100,200])
print(lst)
lst.insert(0,1000)
print(lst)
```

- Item Removing Methods
    - remove(item) : It removes the first occurence item from the list if it is exists in list otherwise `ValueError`
    - pop(pos) : It remove and return the specified position item if position is passed, otherise it remove and returns the last item from list
    - clear(): It removes all elements from list

```python
lst=[10,20,30,20,40,50,20]
lst.remove(20)
print(lst)
#lst.remove(100): ValueError: list.remove(x): x not in list
```

```python
lst=[10,20,30,20,40,50,20]
val=lst.pop()
print(val)
print(lst)
val2=lst.pop(0) # It removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
# index and count()
# index(item): If item is present in list it returns the first occurence Index otherwise `ValueError` 
# count(item): It return the count of Total occurences of item 
lst=[10, 20, 30, 20, 40, 50]
ind=lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# Sorting items
# sort(reverse = False) : By default it sorts the original list into Aescending order.
#                            if reverse is True it sorts in Descending order
lst=[1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
# Sorting items
# sort(reverse = False) : By default it sorts the original list into Aescending order.
#                            if reverse is True it sorts in Descending order
lst=['hi','hello','welcome']
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```
