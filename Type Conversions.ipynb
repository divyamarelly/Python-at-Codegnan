### Type Conversions

```python
# integer to float, boolean, string, complex
num = 10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
```

```python
# float to int, bool and string
val = 10.2556
num = int(val)
print(type(num),num)
print(bool(val))
str_val = str(val)
print(type(str_val),str_val)
print(str_val[:4])
print(str_val[1],str_val[-1])
```

```python
# float to string and printing integer number
val = 102.25545521
str_val = str(val)
ind = str_val.find('.')
print(str_val[:ind])
```

```python
# boolean to int, float, string
b = True
print(int(b))
print(float(b))
print(str(b)[1])
```

```python
# string to int, float bool, list, tuple, set
# if string value is number
s = '123456'
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# string to int, float bool, list, tuple, set
# if string value is float
s = '123456.1205'
print(int(float(s))) # convert to float first next to int
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# string to bool, list, tuple, set
# if string value contains character
s = 'Codegnan'
print(bool(s))
print(list(s))
print(tuple(s))
print(tuple(s)[0:4])
print(set(s))
```

```python
# list to string, tuple, set, boolean
l1 = [1, 4.5, 'Hi', 'Hello', 1,4]
print(str(l1))
print(str(l1)[10:12])
print(tuple(l1))
print(set(l1))
print(bool(l1))
```

```python
# tuple to string, list, set, boolean
t=(1, 2, 3, 10, 23)
print(str(t))
print(str(t)[10:16])
print(list(t))
print(list(t)[3])
print(set(t))
print(bool(t))
```

```python
# set to string, list, tuple
s = {'a', 1, 2, 3, 2.02, 'hi', (10, 20, 30)}
print(str(s))
print(str(s)[16:20])
print(list(s))
print(list(s)[5])
print(tuple(s))
print(tuple(s)[6])
print(bool(s))
```

```python
# dictionary to string, list, tuple, set, boolean
d = {'name': 'Divya', 'course': 'Python', 'batch': 57, 'age': 20}
print(str(d))
print(str(d)[9:16])
print(list(d))
print(tuple(d)[0:2])
print(set(d))
print(bool(d))
```

```python
# list to dictionary conversion
l = [('a',1), ('b',2)]
print(dict(l))
```
