```python
# Types of Inheritance
# Single Inheritance
# child class with constructor
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old and I am a {self.gender}"
        
class Student(Person):
    def __init__(self, stdid, name, gender, age, branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch
        
s = Student(101, "Divya", "Female", 20, "Python")
print(s.name)
print(s.branch)
# print(s.gender) # AttributeError: 'Student' object has no attribute 'gender'
print(s.details())
```

```python
# 2. Multilevel innheritance
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old"
    
class Student(Person):
    def __init__(self, name, age, branch,):
        super().__init__(name, age,)
        self.branch = branch

class Employee(Student):
    def __init__(self, name, age, branch, salary):
        super().__init__(name, age, branch)
        self.salary = salary
    def get_salary(self, bonous=0):
        self.salary += bonous
        return self.salary
        
e = Employee("Divya", 20, "HYD", 50000)
print(e.name)
print(e.branch)
print(e.details())
print(e.get_salary())
print(e.get_salary(bonous = 10000))
```

```python
# 3. Multiple inheritance
class Father:
    def __init__(self, Father_name):
        self.Father_name = Father_name
    def work1(self):
        return "Father working as Private Employee"

class Mother:
    def __init__(self, Mother_name):
        self.Mother_name = Mother_name
    def work2(self):
        return "Mother working as Teacher"

# child class
class Son(Father, Mother):
    def __init__(self, name, age, Father_name, Mother_name):
        Father.__init__(self, Father_name)
        Mother.__init__(self, Mother_name)
        self.name = name
        self.age = age
    def parent_work(self):
        return super().work1() + "\n" + super().work2()

s = Son("Divya", 21, "HanumanthRao", "Dhanalaxmi")
print(s.name)
print(s.Father_name)
print(s.Mother_name)
print(s.parent_work())
```

```python
# 4. Hierarchical Inheritance
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old"

class Student(Person):
    def __init__(self, name, age, branch):
        super().__init__(name, age)
        self.branch = branch
    def student_details(self):
        return f"Branch: {self.branch}"

class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary
    def employee_details(self):
        return f"Salary: {self.salary}"

s = Student("Divya", 20, "Python")
e = Employee("Vamshi", 25, 50000)
print(s.details())
print(s.student_details())
print()
print(e.details())
print(e.employee_details())
```

```python
# Hybrid Inheritance
class Person:
    def __init__(self, name):
        self.name = name
    def details(self):
        return f"Name: {self.name}"

class Student(Person):
    def __init__(self, name, branch):
        Person.__init__(self, name)
        self.branch = branch

class Employee(Person):
    def __init__(self, name, salary):
        Person.__init__(self, name)
        self.salary = salary

class Intern(Student, Employee):
    def __init__(self, name, branch, salary, company):
        Student.__init__(self, name, branch)
        Employee.__init__(self, name, salary)
        self.company = company
    def intern_details(self):
        print(self.details())
        print("Branch:", self.branch)
        print("Salary:", self.salary)
        print("Company:", self.company)

i = Intern("Divya", "Python", 25000, "OpenAI")
i.intern_details()
```
