```python
import re
```

```python
# match(pat, string): It checks whether the string starts with pattern or not
string = "The sky is blue and ocean is blue"
res = re.match("The", string)
print(res)
print(res.span()) # it returns the matched patterns position as tuple
print(res.group()) # it returns the Matched result
```

```python
# match
try:
    string = "The sky is blue and ocean is blue"
    res = re.match("sky", string)
    print(res)
    print(res.span()) # it returns the matched patterns position as tuple
    print(res.group()) # it returns the Matched result
except Exception as e:
    print("There is no Matched pattern")
```

```python
# search(pat, string): It checks the entire string and it returns the first matched pattern
try:
    string = "The sky is blue and ocean is blue"
    res = re.search("blue", string)
    print(res)
    print(res.span()) # it returns the matched patterns position as tuple
    print(res.group()) # it returns the Matched result
except:
    print("There is no Matched pattern")
```

```python
# search(pat, string): It checks the entire string and it returns the first matched pattern
try:
    string = "The sky is blue and ocean is blue"
    res = re.search("z", string)
    print(res)
    print(res.span()) # it returns the matched patterns position as tuple
    print(res.group()) # it returns the Matched result
except:
    print("There is no Matched pattern")
```

```python
# findall(pat, string): It checks the entire string and it returns list of all matched patterns
try:
    string = "The sky is blue and ocean is blue"
    res = re.findall("[abc]", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
# split(pat, string): It splits the String wherever the pattern is matched
try:
    string = "The sky is blue and ocean is blue"
    res = re.split("blue", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
# split(pat, string): It splits the String wherever the pattern is matched
try:
    string = "The sky is blue and ocean is blue"
    res = re.split("[abc]", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
# sub(old_pat, new_pat, string): Replaces the old pattern with new pattern
try:
    string = "The sky is blue and ocean is blue"
    res = re.sub("blue", "Red", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
# findall(pat, string): It checks the entire string and it returns list of all matched patterns
try:
    string = "The sky is blue and ocean is blue"
    res = re.findall("[isabce]+", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
# findall(pat, string): It checks the entire string and it returns list of all matched patterns
try:
    string = "The sky is blue 123 and ocean is bl9ue 2376"
    res = re.findall("[0-9]", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
# validate mobile number
try:
    mobile = input()
    res = re.fullmatch("^[6-9][0-9]{9}",mobile)
    print(res)
    if res:
        print(f"{mobile} is valid")
    else:
        print(f"{mobile} is invalid")
except:
    print("There is no matched pattern")
```

```python
# validate pan number
try:
    pan_no = input()
    res = re.fullmatch("^[A-Z]{5}[0-9]{4}[A-Z]$",pan_no)
    print(res)
    if res:
        print(f"{pan_no} is valid")
    else:
        print(f"{pan_no} is invalid")
except:
    print("There is no matched pattern")
```
