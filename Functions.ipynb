```python
# Types  of functions based on parameters ad return value
# 4. With parameters and with return

# function definition block
def addition(x:int, y:int)->int:
    c = x + y
    return c
# function calling block
a, b = list(map(int, input().split()))
res = addition(a, b)
print(res)
```

```python
# 3. With parameters and without return
# by default function returns None value, when function is not returning any value

# function definition block
def addition(x:int, y:int)->None:
    c = x + y
    print(f"This is from function definition block: {c}")
# function calling block
a, b = 10, 20
res = addition(a, b)
print(f"This is from function calling block: {res}")
```

```python
# 2. Without parameters and with return
# function definition block
def addition()->int:
    x = 10
    y = 20
    z = x + y
    return z
# function calling block
res = addition()
print(f"This is from function calling block: {res}")
```

```python
# 1. Without parameters and without return
# fuction definition block
def addition()->None:
    x = 10
    y = 20
    c = x + y
    print(f"This is from function definition block: {c}")
# function calling block
res = addition()
print(f"This is from function calling block: {res}")
```

```python
# implement even or odd using functions
def even_or_odd(x:int)->str:
    if x % 2 == 0 :
        return "Even"
    else:
        return "Odd"
# function calling block
n = int(input())
print(even_or_odd(n))
```

```python
# prime number using functions
def check_prime(n:int)->str:
    if n < 2:
        return "Not Prime"
    for i in range(2, n//2 + 1):
        if n % i == 0:
            return "Not Prime"
    return "Prime"
#function calling block
num = int(input())
print(check_prime(num))
```
