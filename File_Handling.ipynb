- ### FILE HANDLING

- `File`: The file is used to store data in electronic formate in system.  <br>
   ` File Operations:`  <br> 
      1. Opening a file - open()  <br> 
      2. Reading a file - read()  <br> 
      3. Writing into a file - write()  <br> 
      4. Closing a file - close()  <br> 
 
1. `OPENING A FILE:`  <br> 
            - To open a file by using open() in that, we need to pass filename and file opening mode.  <br> 
                `syntax`:  <br> 
                      file_obj = open(file_name, mode)  <br> 
                           # read/write operations  <br> 
                           # close file.  <br> 
 `FILE OPENING MODEALS:`  <br> 
       r - (read) : If file not exists it rises file not found error.  <br> 
       w - (write) : Opening a file in write mode a file is not exists it creates a new file, otherwise it eareses the previous content.  <br> 
       a - (append) : Opening a file in appened mode a file not exists it creates a new file, otherwise it add content to the previous content.  <br> 
       x - (create : To create a new file. <br> 
       b - (binary) :This mode is used to perform operations on binary files for writing (wb) mode for reading (rb) mode. <br> 
       r+ - (read + write) <br> 
       w+ - (write + read)  <br> 
       a+ - (append + read) <br> 

2. `READING CONTENT FROM FILE:` <br> 
           - To perform raed operation from file for that we need to open in "r" mode and then we can perform follwing operations. <br> 
                    i. read(): It returns entire file content as single string <br> 
                    ii. readline(): This method returns first line. <br> 
                    iii. readlines(): It returns list of lines. <br> 

3. `WRITING INTO A FILE:`  <br> 
           - If you open a file in write mode we can perform following operations. <br> 
                    There are two methods:  <br> 
                       i. write(string): It adds string into a file. <br> 
                       ii. writelines(list_of_lines): It adds all lists lines to a file. <br>

4. `CLOSING A FILE:` <br> 
          - After performing read/write operations on a file, we need to  close the file by using close(), otherwise data lose will happen. <br>
